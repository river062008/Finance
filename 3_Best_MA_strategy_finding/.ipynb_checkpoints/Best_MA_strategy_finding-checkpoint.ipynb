```python
import vectorbt as vbt
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import seaborn as sns
import random


from numba.core.errors import NumbaDeprecationWarning, NumbaPendingDeprecationWarning
import warnings
warnings.simplefilter('ignore', category=NumbaDeprecationWarning)
warnings.simplefilter('ignore', category=NumbaPendingDeprecationWarning)

year = 5
start = datetime.now()-timedelta(days=365*year)
end = datetime.now()
price = vbt.YFData.download(['AAPL', 'MSFT'], start=start, end=end).get('Close')
```

```python
windows = np.arange(5,50,5)
fast_ma, slow_ma = vbt.MA.run_combs(price, window=windows, r=2, short_names=['fast', 'slow'])
entries = fast_ma.ma_crossed_above(slow_ma)
exits = fast_ma.ma_crossed_below(slow_ma)
# group_by = ['first', 'first', 'second']
pf_kwargs = dict(size=np.inf, size_type='Percent', init_cash=10000, fees=0.001, freq='1D', size_granularity=1)
pf = vbt.Portfolio.from_signals(price, entries, exits, **pf_kwargs)
```

```python
pf.orders.records_readable
```

```python
fig = pf.total_return().vbt.heatmap(
    x_level='fast_window', y_level='slow_window', slider_level='symbol', symmetric=True,
    trace_kwargs=dict(colorbar=dict(title='Total return', tickformat='%')))
fig.show()
```
